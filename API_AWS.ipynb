### STEP 1 : API Query

I use census data for employer firms by sector, sex, ethnicity, race, veteran status, and years in business for the U.S., including detailed business characteristics, for the years 2014, 2015 and 2016.

```python
import requests
```

```python
import pandas as pd
```

```python
#Data for 2016
base_url_2016= "https://api.census.gov/data/2016/ase/cscb"

#The Census data website gives a list of all the variables included in a dataset, 
#from which we picked our variables of interest to input as parameters in the API Call
var = """NAME,NAICS2012_LABEL,ASECB,ASECB_LABEL,SPOUSES,SPOUSES_LABEL,YEAR,FIRMPDEMP,FIRMPDEMP_PCT,RCPPDEMP,
       RCPPDEMP_F,RCPPDEMP_PCT,EMP,EMP_PCT,PAYANN,PAYANN_PCT,FIRMPDEMP_S,FIRMPDEMP_PCT_S,RCPPDEMP_S,RCPPDEMP_PCT_S,
       EMP_S,EMP_PCT_S,PAYANN_S,PAYANN_PCT_S"""

#The Census website suggests using ':*' to get data on all the states for our specified country
country = "us:*" 

my_key = "e22f388e634bfb182e14adf23b0485dcdd48180e"

url1 = f"{base_url_2016}?get={var}&for={country}&key={my_key}"
```

```python
r1 = requests.get(url1)
r1.json()
```

```python
#Converting to dataframe
df1 = pd.DataFrame(r1.json())
df1
```

```python
#Set first row of data as Column Names
df1.columns=df1.iloc[0]
df1 = df1[1:]
df1
```

```python
#Data for 2015
base_url_2015= "https://api.census.gov/data/2015/ase/cscb"

#The rest of the variables stay the same as 2016 

url2 = f"{base_url_2015}?get={var}&for={country}&key={my_key}"
```

```python
r2 = requests.get(url2)
r2.json()
```

```python
df2 = pd.DataFrame(r2.json())
df2
```

```python
#Set first row of data as Column Names
df2.columns=df2.iloc[0]
df2 = df2[1:]
df2
```

```python
#Data for 2014
base_url_2014= "https://api.census.gov/data/2014/ase/cscb"

#The column headers for this year's data have slight differences, so we define them again:

var_2014 = "GEO_TTL,NAICS2012_TTL,ASECB,ASECB_TTL,SPOUSES,SPOUSES_TTL,YEAR,FIRMPDEMP,FIRMPDEMP_PCT,RCPPDEMP,RCPPDEMP_F,RCPPDEMP_PCT,EMP,EMP_PCT,PAYANN,PAYANN_PCT,FIRMPDEMP_S,FIRMPDEMP_PCT_S,RCPPDEMP_S,RCPPDEMP_PCT_S,EMP_S,EMP_PCT_S,PAYANN_S,PAYANN_PCT_S"

#The rest of the variables stay the same as 2016

url3 = f"{base_url_2014}?get={var_2014}&for={country}&key={my_key}"
```

```python
r3 = requests.get(url3)
```

```python
r3.json()
```

```python
df3 = pd.DataFrame(r3.json())
df3
```

```python
#Set first row of data as Column Names
df3.columns=df3.iloc[0]
df3 = df3[1:]
df3
```

```python
#Rename Column Names for 2014 to ensure uniformity across all 3 years
df3.rename(columns={'GEO_TTL': 'NAME', 'NAICS2012_TTL': 'NAICS2012_LABEL', 'ASECB_TTL': 'ASECB_LABEL', 'SPOUSES_TTL':'SPOUSES_LABEL'}, inplace = True)
df3
```

```python
frames = [df1, df2, df3]
```

```python
#Final dataframe- Concatenating all 3 years' data
df4 = pd.concat(frames)
df4
```

```python
#Defining functions for database and table creation
def create_database(dbname):
    connection.cursor().execute(f"CREATE DATABASE IF NOT EXISTS {dbname}")
    connection.commit()
    print("DB Created")
    
def create_table(dbname, tablename, formatted_sql_columns):
    connection.cursor().execute(f"USE {dbname}")
    connection.cursor().execute(f"CREATE TABLE IF NOT EXISTS {tablename}({formatted_sql_columns})")
    connection.commit()
    
def delete_table(dbname, tablename):
    connection.cursor().execute(f"USE {dbname}")
    connection.cursor().execute(f"DROP TABLE IF EXISTS {tablename}")
    connection.commit()
```

```python
#Adding data
def add_row(tablename, dbname, cols, row_data):
    esses = ("%s, " * len(cols))[:-2]
    fixed_columns = ",".join(cols)
    sql = f"INSERT INTO {tablename} ({fixed_columns}) VALUES ({esses})"
    connection.cursor().execute(f"USE {dbname}")
    connection.cursor().execute(sql, row_data)
    connection.commit()
    return({"Status": 200})
```
